```python
BOOKS_DATABASE = [
    {"title": "The Silent Patient", "genre": "Mystery", "length": "Medium", "tone": "Suspenseful"},
    {"title": "Gone Girl", "genre": "Mystery", "length": "Long", "tone": "Suspenseful"},
    {"title": "Dune", "genre": "Sci-Fi", "length": "Long", "tone": "Thought-Provoking"},
    {"title": "Project Hail Mary", "genre": "Sci-Fi", "length": "Medium", "tone": "Fast-Paced"},
    {"title": "Atomic Habits", "genre": "Self-Help", "length": "Short", "tone": "Inspiring"},
    {"title": "Deep Work", "genre": "Self-Help", "length": "Short", "tone": "Thought-Provoking"},
    {"title": "The Great Gatsby", "genre": "Fiction", "length": "Short", "tone": "Thought-Provoking"},
    {"title": "To Kill a Mockingbird", "genre": "Fiction", "length": "Medium", "tone": "Inspiring"}
]
```

```python
def get_reader_preferences():
    print("\n--- Welcome To The AI Book Recommendation Engine ---")
    print("Answer a Few Questions To Find Your Next Great Read\n")
    
    genre = input("Choose a Genre (Fiction, Sci-Fi, Mystery, Self-Help): ").strip().lower()
    length = input("Preferred Length? (Short, Medium, Long): ").strip().lower()
    tone = input("What vibe/tone do you want? (Suspenseful, Thought-Provoking, Fast-Paced, Inspiring): ").strip().lower()
    
    return {
        "genre": genre,
          "length": length,
            "tone": tone
            }
```

```python
def match_book(preferences, database):
    recommended_list = []
    for book in database:
        score = 0
        if preferences["genre"] and preferences["genre"] in book["genre"].lower():
            score += 3
        if preferences["tone"] and preferences["tone"] in book["tone"].lower():
            score += 2
        if preferences["length"] and preferences["length"] in book["length"].lower():
            score += 1
        if score > 0:
            recommended_list.append((book, score))
            
    recommended_list.sort(key=lambda x: x[1], reverse=True)
    return recommended_list
```

```python
def recommend_books(results):
    print("\n==========================================")
    print("      YOUR PERSONALIZED READING LIST      ")
    print("==========================================\n")
    
    if not results:
        print("No Exact Matches Found. Try running again with simpler keywords!")
    else:
        for index, (book, score) in enumerate(results, 1):
            match_percentage = int((score / 6) * 100)
            print(f"{index}. '{book['title']}'")
            print(f"   [Details: {book['genre']} | Length: {book['length']} | Tone: {book['tone']}]")
            print(f"   Match Confidence: {match_percentage}%\n")
            
    print("==========================================")
```

```python
if __name__ == "__main__":
    user_preferences = get_reader_preferences()
    matched_books = match_book(user_preferences, BOOKS_DATABASE)
    recommend_books(matched_books)
```
